1. Soma dos dígitos de um número inteiro positivo


```python
def soma_digitos(numero):
    if numero < 10:
        return numero
    return numero % 10 + soma_digitos(numero // 10)

print(soma_digitos(123))  # 6
```

2. Contar quantos dígitos um número inteiro possui

```python
def contar_digitos(numero):
    numero = abs(numero)
    if numero < 10:
        return 1
    return 1 + contar_digitos(numero // 10)

print(contar_digitos(9876))  # 4
```

```python
def inverter_string(texto):
    if texto == "":
        return ""
    return texto[-1] + inverter_string(texto[:-1])

print(inverter_string("recursao"))  # oasrucer
```

4.Verificar se uma string é palíndromo

```python
def eh_palindromo(texto):
    if len(texto) <= 1:
        return True
    if texto[0] != texto[-1]:
        return False
    return eh_palindromo(texto[1:-1])
print(eh_palindromo("arara"))  # True
```

5. Calcular o produto de dois números usando apenas soma

```python
def multiplicar(a, b):
    if b == 0:
        return 0
    if b < 0:
        return -multiplicar(a, -b)
    return a + multiplicar(a, b - 1)

print(multiplicar(3, 4))  # 12
```

6. Encontrar o maior elemento de uma lista

```python
def maior_lista(lista):
    if len(lista) == 1:
        return lista[0]

    maior_do_resto = maior_lista(lista[1:])
    if lista[0] > maior_do_resto:
        return lista[0]
    return maior_do_resto

print(maior_lista([7, 2, 9, 4, 1]))  # 9
```

7. Calcular o máximo divisor comum (MDC) — algoritmo de Euclides

```python
def mdc(a, b):
    if b == 0:
        return a
    return mdc(b, a % b)

print(mdc(48, 18))  # 6
```

8. Contar quantas vezes um elemento aparece em uma lista

```python
def contar_ocorrencias(lista, elemento):
    if lista == []:
        return 0
    if lista[0] == elemento:
        return 1 + contar_ocorrencias(lista[1:], elemento)
    return contar_ocorrencias(lista[1:], elemento)

print(contar_ocorrencias([1, 2, 3, 2, 2, 4], 2))  # 3
```

9. Somar todos os elementos de uma lista

```python
def soma_lista(lista):
    if lista == []:
        return 0
    return lista[0] + soma_lista(lista[1:])

print(soma_lista([1, 2, 3, 4, 5]))  # 15
```

10. Determinar se um número é primo

```python
def verificar_primo(numero, divisor=2):
    if numero < 2:
        return False
    if divisor * divisor > numero:
        return True
    if numero % divisor == 0:
        return False
    return verificar_primo(numero, divisor + 1)

print(verificar_primo(7))   # True
print(verificar_primo(10))  # False
```

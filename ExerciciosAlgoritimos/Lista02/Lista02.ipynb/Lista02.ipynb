```python
# 13. Escreva uma função que recebe uma lista e retorna o maior e o menor valor.
def maior_menor(lista):
    maior = max(lista)
    menor = min(lista)
    return maior, menor

numeros = [5, 8, 2, 9, 1]

maior, menor = maior_menor(numeros)

print("Maior:", maior)
print("Menor:", menor)
```

```python
# 20. Escreva uma função que recebe uma lista de números e retorna uma nova lista com a soma acumulada (ex: [1,2,3] → [1,3,6]).
def soma_acumulada(lista):
    nova_lista = []
    soma = 0

    for num in lista:
        soma = soma + num
        nova_lista.append(soma)

    return nova_lista

numeros = [1, 2, 3]

print(soma_acumulada(numeros))
```

```python
# 26. Crie um programa simples de Lista de Tarefas (TODO List) em Python que funcione no terminal.
tarefas = []

while True:
    print("\n1. Adicionar tarefa")
    print("2. Listar todas as tarefas")
    print("3. Marcar tarefa como concluída")
    print("4. Remover tarefa")
    print("5. Sair")

    try:
        opcao = int(input("Escolha uma opção: "))
    except ValueError:
        print("Digite um número válido.")
        continue

    if opcao == 1:
        nome = input("Digite o nome da tarefa: ")

        tarefa = {
            "nome": nome,
            "feito": False
        }

        tarefas.append(tarefa)

    elif opcao == 2:
        if len(tarefas) == 0:
            print("Nenhuma tarefa cadastrada.")
        else:
            for i, tarefa in enumerate(tarefas, start=1):
                if tarefa["feito"]:
                    status = "FEITO"
                else:
                    status = "ESPERA"

                print(i, "-", tarefa["nome"], "-", status)

    elif opcao == 3:
        if len(tarefas) == 0:
            print("Nenhuma tarefa cadastrada.")
        else:
            try:
                numero = int(input("Digite o número da tarefa: "))

                if 1 <= numero <= len(tarefas):
                    tarefas[numero - 1]["feito"] = True
                    print("Tarefa marcada como concluída.")
                else:
                    print("Número de tarefa inválido.")
            except ValueError:
                print("Digite um número válido.")

    elif opcao == 4:
        if len(tarefas) == 0:
            print("Nenhuma tarefa cadastrada.")
        else:
            try:
                numero = int(input("Digite o número da tarefa para remover: "))

                if 1 <= numero <= len(tarefas):
                    removida = tarefas.pop(numero - 1)
                    print("Tarefa removida:", removida["nome"])
                else:
                    print("Número de tarefa inválido.")
            except ValueError:
                print("Digite um número válido.")

    elif opcao == 5:
        print("Programa encerrado.")
        break

    else:
        print("Opção inválida.")
```
